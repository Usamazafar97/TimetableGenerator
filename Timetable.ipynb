```python
import numpy as np
from numpy import random

# Global Variables
Total_population = 10
Mutation_rate = 3  # 3 percent

# Global Arrays
Teachers = np.array(
    ["Sir Ahmed Nawaz", "Sir Umair Arshad", "Mam Labiba", "Mam Humera", "Mam Amna Irum", "Mam Irum Inayat",
     "Sir Kifayatullah", "Sir Ahsanullah", "Mam Noorul ain", "Mam Amna Basharat", "Sir Ejaz", "Sir Hassan", "Mam Rabia",
     "Mam Sarah", "Sir Ahmed Ali", "Mam Shehr Bano", "Sir Usama Rasheed", "Dr Aqeel Ejaz", "Sir Anas Zafar"])
Courses = np.array(["ITC", "AI", "HCI", "SMD", "TM", "COAL", "CP", "DB", "AP", "CA", "SE", "FYP-I", "FYP-II"])
Sections = np.array(["A", "B", "C", "D", "E", "F", "G"])
Duration = np.array(["0830", "0930", "1030", "1130", "1230", "0130", "0230", "0330", "0430"])
Day = np.array(["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"])
Rooms = np.array([401, 402, 403, 101, 102, 103, 201, 202, 203, 301, 302, 303, 501, 502, 503])

Populations_array = []
# no of days * no. of rooms *no. of hours in a day
# No_of_chromosomes = len(Day) * len(Rooms) * len(Duration)
No_of_chromosomes = 30


# Returning the single gene (Timeslot)
def return_random_slot():
    return [Teachers[random.randint(0, Teachers.size - 1)], Courses[random.randint(0, Courses.size - 1)],
            Sections[random.randint(0, Sections.size - 1)], Duration[random.randint(0, Duration.size - 1)],
            Day[random.randint(0, Day.size - 1)], Rooms[random.randint(0, Rooms.size - 1)]]


# Filling the chromosomes using genes (Timeslot)
def return_choromosomes_array():
    Chromosomes_array = []
    for i in range(No_of_chromosomes):
        gene = return_random_slot()
        Chromosomes_array.append(gene)
    return Chromosomes_array


# Filling the Population with chromosomes (Timetables)
def fill_populations_array():
    for i in range(Total_population):
        chromosome = return_choromosomes_array()
        Populations_array.append(chromosome)


# Printing population array
def printing_population_array():
    fill_populations_array()
    print(Populations_array)


# calclate fitness of the given chromosome
def calculate_fitness(chromosome):
    overall_score = 0

    # chromosome = [
    #     ['Mam Humera', 'ITC', 'D', '0830', 'Wednesday', 304],
    #     ['Mam Humera', 'AI', 'C', '0830', 'Wednesday', 301],
    #     ['Mam Humera', 'COAL', 'F', '0830', 'Tuesday', 303],
    #     ['Sir Umair Arshad', 'COAL', 'A', '1030', 'Tuesday', 203]
    #
    # ]

    boolean_No_teacher_hold_two_classes_same_time = True
    boolean_No_sec_hold_two_classes_same_time = True
    boolean_No_classroom_hold_two_classes_same_time = True
    boolean_No_three_consecutive_classes = True
    boolean_No_three_sections_classes = True
    boolean_No_Friday_1_to_2_classes = True

    for gene in range(len(chromosome)):
        for j in range(gene + 1, len(chromosome)):

            # No teacher can hold two classes at same time
            if (boolean_No_teacher_hold_two_classes_same_time and chromosome[gene][0] == chromosome[j][0] and
                    chromosome[gene][3] == chromosome[j][3]):
                boolean_No_teacher_hold_two_classes_same_time = False
                # print("matched")
                # print(chromosome[gene][0] + "-" + chromosome[j][0])
                # print(chromosome[gene][3] + "-" + chromosome[j][3])

            # No section can listen for two classes at same time
            if (boolean_No_sec_hold_two_classes_same_time and chromosome[gene][1] == chromosome[j][1] and
                    chromosome[gene][2] == chromosome[j][2] and chromosome[gene][3] == chromosome[j][3]):
                boolean_No_sec_hold_two_classes_same_time = False
                # print("matched")
                # print(chromosome[gene][1] + "-" + chromosome[j][1])
                # print(chromosome[gene][2] + "-" + chromosome[j][2])
                # print(chromosome[gene][3] + "-" + chromosome[j][3])

            # No classroom can hold two classes at same time
            if (boolean_No_classroom_hold_two_classes_same_time and chromosome[gene][2] == chromosome[j][2] and
                    chromosome[gene][3] == chromosome[j][3] and chromosome[gene][4] == chromosome[j][4] and
                    chromosome[gene][5] == chromosome[j][5]):
                boolean_No_classroom_hold_two_classes_same_time = False
                # print("matched")
                # print(chromosome[gene][2] + "-" + chromosome[j][2])
                # print(chromosome[gene][3] + "-" + chromosome[j][3])
                # print(chromosome[gene][4] + "-" + chromosome[j][4])
                # print(chromosome[gene][5] + "-" + chromosome[j][5])
                # No classroom can hold two classes at same time

            if (boolean_No_Friday_1_to_2_classes and chromosome[gene][4] == "Friday" and
                    chromosome[gene][3] == "1230" and chromosome[gene][3] == "0130"):
                boolean_No_Friday_1_to_2_classes = False

    #No teacher can take 3 consective classes
    for gene in range(len(chromosome) - 1):
        # teacher = chromosome[gene][0]
        for j in range(gene + 1, len(chromosome) - 1):
            # print(gene, j)
            if boolean_No_three_consecutive_classes and chromosome[gene][0] == chromosome[j][0] and chromosome[gene][
                0] == chromosome[j + 1][0]:
                boolean_No_three_consecutive_classes = False
                # print(chromosome[gene][0] + "-" + chromosome[j][0])

    # No Sec can take 3 consective classes
    for gene in range(len(chromosome) - 1):
        # teacher = chromosome[gene][0]
        for j in range(gene + 1, len(chromosome) - 1):
            # print(gene, j)
            if boolean_No_three_sections_classes and chromosome[gene][2] == chromosome[j][2] and \
                    chromosome[gene][
                        0] == chromosome[j + 1][2]:
                boolean_No_three_sections_classes = False
                # print(chromosome[gene][0] + "-" + chromosome[j][0])

    if (boolean_No_teacher_hold_two_classes_same_time):
        overall_score = overall_score + 1

    if (boolean_No_sec_hold_two_classes_same_time):
        overall_score = overall_score + 1

    if (boolean_No_classroom_hold_two_classes_same_time):
        overall_score = overall_score + 1

    if (boolean_No_three_consecutive_classes):
        overall_score = overall_score + 1

    if (boolean_No_three_sections_classes):
        overall_score = overall_score + 1

    if (boolean_No_Friday_1_to_2_classes):
        overall_score = overall_score + 1

    return overall_score


def roulette_wheel_selection(population_array):
    fitness_array = []
    for i in range(len(population_array)):
        fitness_array.append(calculate_fitness(population_array[i]))
    # print(fitness_array)
    total_sum = sum(fitness_array)

    probability_array = []

    for i in range(len(fitness_array)):
        probability_array.append(fitness_array[i] / total_sum)

    # print(probability_array)

    index_population_array = []
    for i in range(len(population_array)):
        index_population_array.append(i)

    # print(index_population_array)

    returned_index_array = []
    returned_index_array = np.random.choice(index_population_array, 2, probability_array)

    return returned_index_array


def crossover(cromosome_1, cromosome_2):
    # print(cromosome_1)
    # print(cromosome_2)
    random_number = random.randint(0, No_of_chromosomes)
    # print(random_number)

    array_one_part_one_array = []
    array_two_part_one_array = []
    for i in range(random_number):
        array_one_part_one_array.append(cromosome_1[i])
        array_two_part_one_array.append(cromosome_2[i])
    array_one_part_two_array = []
    array_two_part_two_array = []
    for j in range(random_number, No_of_chromosomes):
        array_one_part_two_array.append(cromosome_1[j])
        array_two_part_two_array.append(cromosome_2[j])

    duplicated_cromosomes_1 = []
    duplicated_cromosomes_2 = []

    for i in range(len(array_one_part_one_array)):
        duplicated_cromosomes_1.append(array_one_part_one_array[i])
    for i in range(len(array_two_part_two_array)):
        duplicated_cromosomes_1.append(array_two_part_two_array[i])
    for i in range(len(array_one_part_two_array)):
        duplicated_cromosomes_2.append(array_one_part_two_array[i])
    for i in range(len(array_two_part_one_array)):
        duplicated_cromosomes_2.append(array_two_part_one_array[i])

    return duplicated_cromosomes_1, duplicated_cromosomes_2


def exchange_lowest_fitness(chromosome_1, chromosome_2):
    fitness_array = []
    for i in range(len(Populations_array)):
        fitness_array.append(calculate_fitness(Populations_array[i]))

    lowest = fitness_array[0]
    second_lowest = None
    for i in fitness_array[1:]:
        if i < lowest:
            second_lowest = lowest
            lowest = i
        elif second_lowest == None or second_lowest > i:
            second_lowest = i

    # print("here")
    # print(lowest,second_lowest)
    # print(fitness_array)
    lowest_index = fitness_array.index(lowest)
    second_lowest_index = fitness_array.index(second_lowest)

    # print(lowest_index,second_lowest_index)
    #
    # print("chromosome_1",calculate_fitness(chromosome_1))
    # print("chromosome_2",calculate_fitness(chromosome_2))
    # print("lowest_index",calculate_fitness(Populations_array[lowest_index]))
    # print("second_lowest_index",calculate_fitness(Populations_array[second_lowest_index]))

    if (calculate_fitness(chromosome_1) > lowest):
        # print("calculate_fitness(Populations_array[lowest_index]) < lowest")
        Populations_array[lowest_index] = chromosome_1
        # print("before mutation",Populations_array[lowest_index])
        mutation(Populations_array[lowest_index])
        # print("after mutation", Populations_array[lowest_index])
    if (calculate_fitness(chromosome_2) > second_lowest):
        # print("calculate_fitness(Populations_array[second_lowest_index]) < second_lowest")
        Populations_array[second_lowest_index] = chromosome_2
        # print("before mutation", Populations_array[lowest_index])
        mutation(Populations_array[second_lowest_index])
        # print("after mutation", Populations_array[lowest_index])


def mutation(chromosome):
    rand_mutation_rate = random.randint(0, 99999) % 100
    if rand_mutation_rate > Mutation_rate:
        return

    random_gene_index = random.randint(0, len(chromosome))
    random_duration = Duration[random.randint(0, Duration.size - 1)]

    chromosome[random_gene_index][3] = random_duration


def return_best_chromosome():
    fitness_array = []
    for i in range(len(Populations_array)):
        fitness_array.append(calculate_fitness(Populations_array[i]))
    maximum = max(fitness_array)
    max_index = fitness_array.index(maximum)

    return Populations_array[max_index]


def main():
    generation = 0
    fill_populations_array()
    while (True):
        print(generation)
        print(calculate_fitness(return_best_chromosome()))
        if (calculate_fitness(return_best_chromosome()) >= 6):
            best_chromosome = return_best_chromosome()
            for row in best_chromosome:
                for elem in row:
                    print(elem, end=' ')
                print()
            break

        n_childrens = 6

        for i in range(n_childrens // 2):
            returnedindexes = roulette_wheel_selection(Populations_array)
            c1, c2 = crossover(Populations_array[returnedindexes[0]], Populations_array[returnedindexes[1]])
            exchange_lowest_fitness(c1, c2)
        generation += 1



main()
```
